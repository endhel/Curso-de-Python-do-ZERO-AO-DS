```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv('datasets/aula02.csv')
df.head()
```

```python
# Quantos imoveis existem por numero de quartos
df[['id','bedrooms']].groupby('bedrooms').count()
```

# Respondendo das perguntas do CEO

```python
# 1 - Qual e o numero de imoveis por ano de construcao?
df[['id', 'yr_built']].groupby('yr_built').count().reset_index()
```

```python
# 2 - Qual e o menor numero de quartos por ano de construcao?
df[['bedrooms', 'yr_built']].groupby('yr_built').min().reset_index()
```

```python
# 3 - Qual e o preco de compra mais alto por numero de quarto?
df[['bedrooms', 'price']].groupby('bedrooms').max().reset_index()
```

```python
# 4 - Qual e a soma de todos precos de compra por numero de quarto?
df[['bedrooms', 'price']].groupby('bedrooms').sum().reset_index()
```

```python
# 5 - Qual e a soma de todos precos de compra por numero de quarto e banheiro?
df[['bedrooms', 'bathrooms', 'price']].groupby(['bedrooms', 'bathrooms']).sum().reset_index()
```

```python
# 6 - Qual e o tamanho medio das salas por ano de construcao?
df[['sqft_living', 'yr_built']].groupby('yr_built').mean().reset_index()
```

```python
# 7 - Qual e o tamanho mediano das salas por ano de construcao?
df[['sqft_living', 'yr_built']].groupby('yr_built').median().reset_index()
```

```python
# 8 - Qual e o desvio padrao do tamanho das salas por ano de construcao?
df[['sqft_living', 'yr_built']].groupby('yr_built').std().reset_index()
```

```python
# 9 - Qual e o crescimento total dos precos de compras dos imoveis, por ano, dia e semana do ano?

#First graph

import matplotlib.pyplot as plt

df['year'] = pd.to_datetime(df.date).dt.year
by_year = df[['price', 'year']].groupby('year').sum().reset_index()

plt.figure( figsize=(20, 12))
plt.bar(by_year.year, by_year.price)
```

```python
#Second graph

df['day'] = pd.to_datetime(df.date)
by_day = df[['price', 'day']].groupby('day').mean().reset_index()

plt.figure( figsize=(20, 12))
plt.plot(by_day.day, by_day.price)
```

```python
#third graph

df['year_week'] = pd.to_datetime(df.date).dt.strftime('%Y-%U')
by_year_week = df[['price', 'year_week']].groupby('year_week').mean().reset_index()

plt.figure( figsize=(20, 12))
plt.plot(by_year_week.year_week, by_year_week.price)
plt.xticks(rotation=60);
```

```python
# Dashboard

from matplotlib import gridspec

fig = plt.figure( figsize=(20, 12))
specs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1 = fig.add_subplot(specs[0, :])
ax2 = fig.add_subplot(specs[1, 0])
ax3 = fig.add_subplot(specs[1, 1])

df['year'] = pd.to_datetime(df.date).dt.year
by_year = df[['price', 'year']].groupby('year').sum().reset_index()
ax1.bar(by_year.year, by_year.price)

df['day'] = pd.to_datetime(df.date)
by_day = df[['price', 'day']].groupby('day').mean().reset_index()
ax2.plot(by_day.day, by_day.price)

df['year_week'] = pd.to_datetime(df.date).dt.strftime('%Y-%U')
by_year_week = df[['price', 'year_week']].groupby('year_week').mean().reset_index()
ax3.plot(by_year_week.year_week, by_year_week.price)
plt.xticks(rotation=60);
```
